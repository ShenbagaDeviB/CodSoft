```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('fraudTrain.csv')
train
```

```python
test = pd.read_csv('fraudTest.csv')
test
```

```python
df = pd.concat([train,test])
df
```

```python
df.info()
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df['trans_date_trans_time']=label_encoder.fit_transform(df['trans_num'])
df['merchant']=label_encoder.fit_transform(df['merchant'])
df['category']=label_encoder.fit_transform(df['category'])
df['first']=label_encoder.fit_transform(df['first'])
df['last']=label_encoder.fit_transform(df['last'])
df['gender']=label_encoder.fit_transform(df['gender'])
df['street']=label_encoder.fit_transform(df['street'])
df['city']=label_encoder.fit_transform(df['city'])
df['state']=label_encoder.fit_transform(df['state'])
df['job']=label_encoder.fit_transform(df['job'])
df['dob']=label_encoder.fit_transform(df['dob'])
df['trans_num']=label_encoder.fit_transform(df['trans_num'])
df['is_fraud']=label_encoder.fit_transform(df['is_fraud'])
print(df.info())
```

```python
df.drop(["trans_date_trans_time","dob"],axis=1)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,df.columns!='is_fraud']
y = df.iloc[:,df.columns=='is_fraud']
```

```python
print(y)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3)
```

```python
xtrain.shape
ytrain.head
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(xtrain, ytrain)
```

```python
predict_output = model.predict(xtest)
print(predict_output)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(predict_output,ytest)
print('The accuracy score for RF:',acc)
```
